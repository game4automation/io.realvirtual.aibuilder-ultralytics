```python
import matplotlib.pyplot as plt
```

```python
def show_single_sample_obb(sample, annotation):
        
    img = sample
    img = img[..., [2, 1, 0]]

    print(img.shape)

    plt.imshow(img)
    plt.axis('off')

    #plt.title('Sample ' + str(i))

    ry, rx = sample.shape[:2]
    # foreach row in annotations[i] draw the obb
    for row in annotation:
        label, x1, y1, x2, y2, x3, y3, x4, y4 = row

        x1 *= rx
        x2 *= rx
        x3 *= rx
        x4 *= rx

        y1 *= ry
        y2 *= ry
        y3 *= ry
        y4 *= ry

        print(row)
            
        # use colors from matplotlib
        color = plt.cm.Spectral((label % 10)/10.0)

        plt.plot([x1,x2,x3,x4,x1],
                [y1,y2,y3,y4,y1],
                color=color)

    plt.show()
```

```python
step = len(samples) // 20
for i in range(20):
    idx = i
    show_single_sample_obb(samples[idx], annotations[idx])
```
