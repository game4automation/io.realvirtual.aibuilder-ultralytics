# Inference

```python
from ultralytics import YOLO
import torch
```

```python
model_name = 'lego_urp_1k'
```

```python
run_folder = 'run'
```

```python
model_path = f'{run_folder}/{model_name}/weights/best.pt'
model = YOLO(model_path)
model.device
```

```python
if torch.cuda.is_available():
    model.to('cuda')

model.device
```

```python
import glob



def predict_folder(folder, imgsz, conf):
    for image_path in glob.glob(folder + '/*'):
        predict_file(image_path, imgsz, conf)

    

def predict_file(image_path, imgsz, conf):
    results = model(
        image_path,
        imgsz=imgsz,  # default is 640
        conf=conf  # default is 0.25
    )
    rvy.show_results(results, model)
        
```
